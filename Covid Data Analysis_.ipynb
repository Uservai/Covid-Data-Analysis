```python
print("Number of persons state wise vaccinated for first dose in India")
first_dose = covid.groupby('State')[['First Dose Administered']].sum()
first_dose
```

```python
print("Number of persons state wise vaccinated for second dose in India")

first_dose = covid.groupby('State')[['Second Dose Administered']].sum()
first_dose
```
